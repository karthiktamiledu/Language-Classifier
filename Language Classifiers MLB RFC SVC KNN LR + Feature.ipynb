```python
import pandas as pd
import numpy as np
import time as time

from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split

from sklearn.metrics import classification_report,confusion_matrix

from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import SelectPercentile
from sklearn.feature_selection import chi2
from sklearn.feature_selection import VarianceThreshold

from sklearn.naive_bayes import MultinomialNB
model_MNB = MultinomialNB()
from sklearn.ensemble import RandomForestClassifier
model_RFC = RandomForestClassifier()
from sklearn import svm
model_SVC = svm.SVC()
from sklearn.neighbors import KNeighborsClassifier
model_KNN = KNeighborsClassifier()
from sklearn.linear_model import LogisticRegression
model_LRC = LogisticRegression()
```

```python
data = pd.read_csv("dataset.csv")
print(data.head())
```

```python
x = np.array(data["Text"])
y = np.array(data["language"])

cv = CountVectorizer()
X = cv.fit_transform(x)
X = SelectPercentile(chi2, percentile = 10).fit_transform(X, y)
X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                    test_size = 0.33,
                                                    random_state = 42)
```

```python
train_avg_MNB = 0
for i in range(19):
    start = time.time()
    model_MNB.fit(X_train, y_train)
    end = time.time()
    train_avg_MNB += (end-start)
train_avg_MNB /= 20
print("train: ", train_avg_MNB)
```

```python
test_avg_MNB = 0
for i in range(19):
    start = time.time()
    model_MNB.predict(X_test)
    end = time.time()
    test_avg_MNB += (end-start)
test_avg_MNB /= 20
print("test average time: ", test_avg_MNB)
```

```python
print(classification_report(y_test, model_MNB.predict(X_test)))
```

```python
model_MNB.fit(X_train, y_train)
model_MNB.predict(X_test)
```
